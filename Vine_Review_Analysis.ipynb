```python
# Dependencies and Setup
import pandas as pd
```

```python
# File to Load 
vine_table_path= "vine_table.csv"
```

```python
#Data Frame
vine_table_df = pd.read_csv(vine_table_path)
vine_table_df
```

```python
vine_table_df
```

```python
# Filter the data and create a new DataFrame to retrieve all the rows where the total_votes count is equal to or greater than 20 
vine_table_df=vine_table_df[vine_table_df["total_votes"]>=20]
vine_table_df
```

```python
#Filter the new DataFrame to retrieve all the rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.

helpfulReview_s = vine_table_df["helpful_votes"] / vine_table_df["total_votes"] >= 0.5

vine_helpfulReview_df = vine_table_df[helpfulReview_s]
vine_helpfulReview_df
```

```python
#Filter the DataFrame to retrieves all the rows where a review was written as part of the Vine program (paid),
paid_s = vine_helpfulReview_df["vine"] == 'Y'

paid_df = vine_helpfulReview_df[paid_s]
paid_df
```

```python
vine_helpfulReview_df["vine"].value_counts()
```

```python
# #Filter the DataFrame to retrieves all the rows where a review was written as part of the Vine program (unpaid),
unpaid_df = vine_helpfulReview_df[~paid_s]
unpaid_df
```

```python
# Nulls
vine_table_df["review_id"].value_counts()
```

```python
# Any nulls in DF
value = vine_helpfulReview_df.isnull().values.any()
value 
```

```python
# Any nulls in review id
value = vine_helpfulReview_df[['review_id']].isnull().values.any()
print(value)
```

```python
# Determine the total number of reviews.
total_reviews = vine_helpfulReview_df[("review_id")].count()
total_reviews
```

```python
# Determine the total number of paid reviews.
total_paidReviews = paid_df[("review_id")].count()
total_paidReviews
```

```python
# Determine the total number of un-paid reviews.
total_unpaidReviews = unpaid_df[("review_id")].count()
total_unpaidReviews
```

```python
# Determine the total number of 5 star reviews.
total_5StarReviews = vine_helpfulReview_df[(vine_helpfulReview_df["star_rating"] == 5)].count()["review_id"]
total_5StarReviews
```

```python
# Determine the total number of paid 5 star reviews.
total_5StarPaidReviews = paid_df[(paid_df["star_rating"] == 5)].count()["review_id"]
total_5StarPaidReviews
```

```python
# Determine the total number of un-paid 5 star reviews.
total_5StarUnPaidReviews = unpaid_df[(unpaid_df["star_rating"] == 5)].count()["review_id"]
total_5StarUnPaidReviews
```

```python
# Determine the  percentage of 5-star reviews paid 5 star reviews.
percentage_paid_reviews= (total_5StarPaidReviews/total_paidReviews)*100
percentage_paid_reviews
```

```python
# Determine the  percentage of 5-star reviews un paid 5 star reviews.
percentage_unpaid_reviews= (total_5StarUnPaidReviews/total_unpaidReviews)*100
percentage_unpaid_reviews
```

```python
#percentage total paid reviews
total_paidReviews_perc= (total_paidReviews/total_reviews)*100
total_paidReviews_perc
```

```python
#percentage total unpaid reviews
total_unpaidReviews_perc= (total_unpaidReviews/total_reviews)*100
total_unpaidReviews_perc
```
